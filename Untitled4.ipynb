```python
# 技能資料
skills = {
    "火球術": {"mp_cost": 10, "damage": 30, "branch": "火焰"},
    "烈焰爆": {"mp_cost": 20, "damage": 60, "branch": "火焰"},
    "寒冰箭": {"mp_cost": 10, "damage": 25, "branch": "冰霜"},
    "冰封世界": {"mp_cost": 25, "damage": 50, "branch": "冰霜"},
}

# 商店資料
shop_items = {
    "回復藥": {"type": "item", "price": 20},
    "鋼劍": {"type": "weapon", "atk": 5, "price": 50},
    "皮甲": {"type": "armor", "def": 4, "price": 40}
}

# 掉落資料
enemy_drops = {
    "史萊姆": ["回復藥", "黏液球"],
}
class Equipment:
    def __init__(self, name, atk=0, defense=0):
        self.name = name
        self.atk = atk
        self.defense = defense
class Character:
    def __init__(self, name, hp, atk, defense, mp, job="冒險者"):
        self.name = name
        self.max_hp = hp
        self.hp = hp
        self.base_atk = atk
        self.base_def = defense
        self.mp = mp
        self.level = 1
        self.exp = 0
        self.gold = 100
        self.skills = ["火球術"]
        self.skill_cd = {}
        self.weapon = None
        self.armor = None
        self.bag = []
        self.job = job
        self.stat_points = 5
        self.str = 5
        self.vit = 5
        self.int = 5
        self.enemy_kills = {}

    def is_alive(self):
        return self.hp > 0

    def atk(self):
        return self.base_atk + (self.weapon.atk if self.weapon else 0) + self.str

    def defense(self):
        return self.base_def + (self.armor.defense if self.armor else 0) + self.vit
def attack(attacker, defender):
    dmg = max(attacker.atk() - defender.defense(), 1)
    defender.hp -= dmg
    print(f"{attacker.name} 攻擊 {defender.name}，造成 {dmg} 傷害")

def use_skill(user, target, skill_name):
    if skill_name not in skills:
        print("技能不存在")
        return
    skill = skills[skill_name]
    if user.mp >= skill["mp_cost"] and user.skill_cd.get(skill_name, 0) == 0:
        user.mp -= skill["mp_cost"]
        target.hp -= skill["damage"]
        user.skill_cd[skill_name] = 3
        print(f"{user.name} 使用 {skill_name} 對 {target.name} 造成 {skill['damage']} 傷害")
    else:
        print("技能冷卻中或 MP 不足")

def reduce_cd(character):
    for skill in character.skill_cd:
        if character.skill_cd[skill] > 0:
            character.skill_cd[skill] -= 1

def battle(player, enemy):
    print(f"你遇到 {enemy.name}！")
    while player.is_alive() and enemy.is_alive():
        print(f"\n{player.name} HP: {player.hp}/{player.max_hp}, MP: {player.mp}")
        print(f"{enemy.name} HP: {enemy.hp}")
        action = input("1. 普通攻擊  2. 技能\n")
        if action == "1":
            attack(player, enemy)
        elif action == "2":
            print("可用技能：", player.skills)
            skill = input("輸入技能名稱：")
            use_skill(player, enemy, skill)
        else:
            print("無效輸入")

        if enemy.is_alive():
            attack(enemy, player)

        reduce_cd(player)

    if player.is_alive():
        print(f"{enemy.name} 被打倒了！")
        drop_loot(player, enemy.name)
    else:
        print("你被打倒了...")
quests = {
    "討伐10隻史萊姆": {"target": "史萊姆", "count": 10, "reward": 100},
}

def check_quests(player):
    for quest, detail in list(quests.items()):
        killed = player.enemy_kills.get(detail["target"], 0)
        if killed >= detail["count"]:
            print(f"完成任務：{quest}，獲得 {detail['reward']} G！")
            player.gold += detail["reward"]
            del quests[quest]
def open_shop(player):
    print("--- 商店 ---")
    for i, (name, item) in enumerate(shop_items.items(), 1):
        print(f"{i}. {name} - {item['price']} G")
    choice = input("輸入編號購買（q 離開）：")
    if choice == 'q': return
    index = int(choice) - 1
    name = list(shop_items.keys())[index]
    item = shop_items[name]
    if player.gold >= item['price']:
        player.gold -= item['price']
        if item["type"] == "weapon":
            player.weapon = Equipment(name, atk=item["atk"])
        elif item["type"] == "armor":
            player.armor = Equipment(name, defense=item["def"])
        else:
            player.bag.append(name)
        print(f"購買成功：{name}")
    else:
        print("金幣不足")
job_data = {
    "戰士": {"str": 3, "vit": 2},
    "法師": {"int": 4, "mp": 10},
}

def choose_job(player):
    print("選擇職業：1. 戰士 2. 法師")
    job = input()
    if job == "1":
        player.job = "戰士"
        player.str += job_data["戰士"]["str"]
        player.vit += job_data["戰士"]["vit"]
    elif job == "2":
        player.job = "法師"
        player.int += job_data["法師"]["int"]
        player.mp += job_data["法師"]["mp"]
import random

def explore(player):
    event = random.choice(["金幣", "敵人", "裝備", "什麼都沒有"])
    if event == "金幣":
        gold = random.randint(20, 50)
        player.gold += gold
        print(f"你找到 {gold} G！")
    elif event == "敵人":
        enemy = Character("史萊姆", 30, 5, 2, 0)
        battle(player, enemy)
    elif event == "裝備":
        loot = Equipment("短劍", atk=3)
        player.weapon = loot
        print("你撿到一把短劍！")
    else:
        print("探索失敗，什麼都沒找到。")
def drop_loot(player, enemy_name):
    loot = random.choice(enemy_drops.get(enemy_name, []))
    player.bag.append(loot)
    player.enemy_kills[enemy_name] = player.enemy_kills.get(enemy_name, 0) + 1
    print(f"{enemy_name} 掉落了：{loot}")
def forge(player):
    if "黏液球" in player.bag:
        player.bag.remove("黏液球")
        if player.weapon:
            player.weapon.atk += 2
            print("武器強化成功！攻擊力 +2")
        else:
            print("你沒有武器")
    else:
        print("沒有強化材料")
def allocate_stats(player):
    while player.stat_points > 0:
        print(f"剩餘點數：{player.stat_points}")
        stat = input("分配到 str / vit / int：")
        if stat == "str":
            player.str += 1
        elif stat == "vit":
            player.vit += 1
        elif stat == "int":
            player.int += 1
        else:
            print("輸入錯誤")
        player.stat_points -= 1
def main():
    hero = Character("勇者", 100, 10, 5, 30)
    choose_job(hero)

    while hero.is_alive():
        print("\n=== 主選單 ===")
        print("1. 探索")
        print("2. 商店")
        print("3. 任務")
        print("4. 鍛造")
        print("5. 分配屬性點")
        print("6. 查看背包")
        print("7. 離開")

        choice = input("輸入選項：")
        if choice == "1":
            explore(hero)
        elif choice == "2":
            open_shop(hero)
        elif choice == "3":
            check_quests(hero)
        elif choice == "4":
            forge(hero)
        elif choice == "5":
            allocate_stats(hero)
        elif choice == "6":
            print("背包內容：", hero.bag)
        elif choice == "7":
            print("遊戲結束。")
            break
        else:
            print("無效輸入")
main()
```
